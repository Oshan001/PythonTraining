```python
import matplotlib.pyplot as plt
label = ["oshan","dssdf",'sdsdfs','dgsg']
man_age=[45,55,47,22]
woman_age=[55,55,5,5]

```

```python
fig,ax=plt.subplots()
ax.bar(label,man_age,label="men")
ax.bar(label,woman_age,label="women")
ax.set_ylabel("Age")
ax.set_ylabel("Group number")
ax.set_ylabel("Gender ratio")
ax.legend()
plt.show()
```

```python
import numpy as np
classes=["python","r","a1","ML","MS"]
class1_students=[30,10,20,26,10]
class2_student = [40,5,20,20,10]
class_index=np.arange(len(classes))
width=0.2
plt.bar(class_index,class1_students,width, label="class1_student")
plt.bar(class_index,class2_student,width, label="class2_student")
plt.xticks(class_index + width,classes)
plt.title("bar chat of classes")
plt.xlabel("classes")
plt.ylabel("no of student")
plt.show()
```

```python
x_axis = np.array([0,5,10,15,20,25,30])
plt.plot(x_axis,x_axis*2)
plt.xlabel("X axis")
plt.ylabel("y axis")
plt.show()
```

```python
plt.plot(x_axis,x_axis*2,"red",label='x=y**2')
plt.legend()
plt.show()
```

```python
y_axis = x_axis**2
empty_canvas = plt.figure()
create_axis = empty_canvas.add_axes([0, 0, 1, 1]) 

create_axis.plot(x_axis, y_axis, 'b')



create_axis = empty_canvas.add_axes([0, 0, 1, 1]) 

create_axis.plot(x_axis, y_axis, 'b', label = "x = x**2")
create_axis.plot(x_axis, y_axis+(x_axis*3), 'g', label = "x = y+(x*3)")
  
create_axis.legend(loc=0)
```
